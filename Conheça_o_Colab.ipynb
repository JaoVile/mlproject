```
# Importar as bibliotecas necess√°rias
!pip install gdown --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")

# Arquivo via Gdown
import gdown

# ID do arquivo no  Drive
file_id = "1IVQKXVyNYtOeoKlk9bohVE5KaTXKd94H"


# URL de download formatada
url = f"https://drive.google.com/uc?id={file_id}"
output_filename = "dataset.csv"

gdown.download(url, output_filename, quiet=False)

# Carregar o dataset
df = pd.read_csv(output_filename)


# Explora√ß√£o inicial
print("Shape:", df.shape)
print(df.dtypes)
print(df.head(10))
print(df.info())
print(df.describe(include='all'))

# Verificar valores nulos e duplicados
print("\nValores nulos por coluna:")
print(df.isnull().sum())
print("\nN√∫mero de duplicatas:", df.duplicated().sum())



# Limpeza b√°sica
df = df.drop_duplicates()





# Aqui voc√™ pode fazer tratamento de nulos conforme seu dataset
# Exemplo gen√©rico:
# df = df.dropna()  # cuidado: s√≥ se fizer sentido

# Convers√£o de tipos se necess√°rio (exemplo)
# df['sua_coluna_data'] = pd.to_datetime(df['sua_coluna_data'], errors='coerce')

# Visualiza√ß√µes




## Distribui√ß√£o de uma vari√°vel num√©rica (ajuste o nome da coluna)
col_num = 'nome_da_coluna_numerica'  # ajuste para sua coluna
if col_num in df.columns:
    plt.figure(figsize=(8,6))
    sns.histplot(df[col_num].dropna(), bins=30, kde=True, color='skyblue')
    plt.title(f'Distribui√ß√£o de {col_num}')
    plt.xlabel(col_num)
    plt.ylabel('Frequ√™ncia')
    plt.show()









## Contagem de uma vari√°vel categ√≥rica (ajuste o nome da coluna)
col_cat = 'nome_da_coluna_categorica'  # ajuste para sua coluna
if col_cat in df.columns:
    plt.figure(figsize=(8,6))
    sns.countplot(data=df, x=col_cat, palette='viridis')
    plt.title(f'Contagem de {col_cat}')
    plt.xlabel(col_cat)
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)
    plt.show()








## Boxplot para verificar outliers numa coluna num√©rica
if col_num in df.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col_num].dropna())
    plt.title(f'Boxplot de {col_num}')
    plt.show()










## Scatterplot entre duas vari√°veis (ajuste os nomes)
x_var = 'coluna_x'  # ajuste
y_var = 'coluna_y'  # ajuste
if x_var in df.columns and y_var in df.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(data=df, x=x_var, y=y_var, alpha=0.7)
    plt.title(f'{x_var} vs {y_var}')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.show()












## Matriz de correla√ß√£o entre vari√°veis num√©ricas
plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=['float64','int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correla√ß√£o')
plt.show()









# Insights iniciais
print("\nüöÄ Insights iniciais sugeridos:")
print("- Verifique quais colunas t√™m muitos valores nulos (acima de 50% por exemplo).")
print("- Verifique quais vari√°veis num√©ricas t√™m fortes correla√ß√µes positivas ou negativas (> |0.7|).")
print("- Verifique domina√ß√£o de categorias numa vari√°vel categ√≥rica.")
print("- Verifique presen√ßa de outliers que podem distorcer a an√°lise.")
print("- Avalie se alguma coluna precisa de transforma√ß√£o (log, escala, etc).")
```
